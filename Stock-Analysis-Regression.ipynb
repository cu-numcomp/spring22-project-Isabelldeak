$\textbf{Description}$

Stock Analysis uses linear regression to perform an analysis on stock prices and the program also displays multiply graphs that plot the stocks prices. It works by getting the data from Yahoo Finance api and then performing different regression implementations on the close price data. The program plots multiple graphs such as the linear regression graph as well as a trendline graph for the stock prices. The program is used to predict the stock prices for the next day. This program is a general purpose software written in python. 




In addition to linear regression model it also uses Support Vector Regression(SVR) to predict future stock prices. Support Vector Regression(SVR)is a supervised learning algorithm that is used to predict discrete values. SVR basically find the best-fit line that is required to fid the data which is called hyperplane. SVR uses kernels which are mathematical functions that takes data as input and transform it into the required form. These are used to find a hyperplane in the higher dimensional space. In this program the contributors are using the following kernels: Linear, Polynomial, and Radial Basis Function (RBF). SVR is different from linear regression model because instead of trying to minimize the error between the actual and predicted value, the SVR tries to fit the best line the hyperplane and the boundary line.

$\textbf{Metrics and Features}$

The code owner tested out his the program to figure out how accurate it is. He did this by testing it against a popular machine learning and regression api called sci-kit-learn. He compared the results and found out that his program produced a similar result but more accurate than the result that was produced by the sci-kit-learn. He measures accuracy as having greater significant figures. His program outputs a number with a greater significant figures which is why it would be slightly more accurate than the result producted by sci-kit-learn. 

When running the program for the stock Nvidia the results were very accurate. The reason the program is accurate is because they use a linear regression model which fits a straight line that minimizes the difference between predicted and actual output values.
